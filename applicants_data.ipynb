```python
import pandas as pd 
df['cv_submission_date'] = pd.to_datetime(df['cv_submission_date'], format='%d-%m-%Y', errors='coerce')
df['interview_date'] = pd.to_datetime(df['interview_date'], format='%d-%m-%Y', errors='coerce')
df['hired_date'] = pd.to_datetime(df['hired_date'], format='%d-%m-%Y', errors='coerce')
df['dob'] = pd.to_datetime(df['dob'], format='%d-%m-%Y', errors='coerce')
```

```python
df.info()
```

```python
df['salary_offered'] = pd.to_numeric(df['salary_offered'], errors='coerce')
```

```python
df['email_id'] = df['email_id'].astype('string')
```

```python
cat_cols = ['applicant_name', 'recruiter_name' ,'process', 'sources', 'stages', 'status', 'gender']
df[cat_cols] = df[cat_cols].astype('category')
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'process')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot size
sns.set(rc={'figure.figsize': (20, 6)})

# Plot
ax = sns.countplot(data=df, x='process', hue='status', palette='Set2')

# Rotate x-axis labels safely
plt.xticks(rotation=45)

# Add count labels to each bar
for bars in ax.containers:
    ax.bar_label(bars)

# Add titles and labels
plt.title("Applicant Count by Process and Status")
plt.xlabel("Process")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the data
df = pd.read_csv('F:\\python\\New folder\\applicant_data.csv', encoding='unicode_escape')

# Total number of applications
total_application = len(df)

# Number of hired applicants
hired_count = df[df['stages'].str.lower() == 'hired'].shape[0]

# Calculate conversion rate
conversion_rate = round((hired_count / total_application) * 100, 2)

# Output
print("Total Applications:", total_application)
print("Hired Applicants:", hired_count)
print("Conversion Rate (%):", conversion_rate)
```
